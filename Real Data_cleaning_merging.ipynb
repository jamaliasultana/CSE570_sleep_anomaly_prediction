```python
# Removing columns from the three CSVs containing 'weekend', 'weekday', 'dis', or 'norm'

# Sleep CSV
columns_to_drop = sleep_csv_10_participants.columns[sleep_csv_10_participants.columns.str.contains('weekend|weekday|dis|norm', case=False)]
sleep_csv_10_participants_cleaned = sleep_csv_10_participants.drop(columns=columns_to_drop)

# Dropping the columns with more than 50% null or NaN values
threshold = len(sleep_csv_10_participants_cleaned) * 0.5
sleep_csv_10_participants_cleaned = sleep_csv_10_participants_cleaned.dropna(axis=1, thresh=threshold + 1)

# Print the remaining number of columns
print(len(sleep_csv_10_participants_cleaned.columns))

# Steps CSV
columns_to_drop = steps_csv_10_participants.columns[steps_csv_10_participants.columns.str.contains('weekend|weekday|dis|norm', case=False)]
steps_csv_10_participants_cleaned = steps_csv_10_participants.drop(columns=columns_to_drop)

# Dropping the columns with more than 50% null or NaN values
threshold = len(steps_csv_10_participants_cleaned) * 0.5
steps_csv_10_participants_cleaned = steps_csv_10_participants_cleaned.dropna(axis=1, thresh=threshold + 1)

# Print the remaining number of columns
print(len(steps_csv_10_participants_cleaned.columns))

# Screen CSV
columns_to_drop = screen_csv_10_participants.columns[screen_csv_10_participants.columns.str.contains('weekend|weekday|dis|norm', case=False)]
screen_csv_10_participants_cleaned = screen_csv_10_participants.drop(columns=columns_to_drop)

# Dropping the columns with more than 50% null or NaN values
threshold = len(screen_csv_10_participants_cleaned) * 0.5
screen_csv_10_participants_cleaned = screen_csv_10_participants_cleaned.dropna(axis=1, thresh=threshold + 1)

# Print the remaining number of columns
print(len(screen_csv_10_participants_cleaned.columns))
```

```python
# Memory optimization function
def optimize_memory_usage(df):
    for col in df.select_dtypes(include=['float64']).columns:
        df[col] = pd.to_numeric(df[col], downcast='float')
    for col in df.select_dtypes(include=['int64']).columns:
        df[col] = pd.to_numeric(df[col], downcast='integer')
    return df

# Apply memory optimization
sleep_csv_10_participants_cleaned = optimize_memory_usage(sleep_csv_10_participants_cleaned)
steps_csv_10_participants_cleaned = optimize_memory_usage(steps_csv_10_participants_cleaned)
screen_csv_10_participants_cleaned = optimize_memory_usage(screen_csv_10_participants_cleaned)
```

```python
# Check for duplicated columns before concatenation
def check_and_remove_duplicates(df1, df2):
    common_columns = [col for col in df1.columns if col in df2.columns and col != 'pid']
    if common_columns:
        print(f"Found common columns: {common_columns}")
        # Drop the common columns from df2 (except 'pid')
        df2 = df2.drop(columns=common_columns)
    return df2

# Ensure no duplicate columns are present before concatenation
steps_csv_10_participants_cleaned = check_and_remove_duplicates(sleep_csv_10_participants_cleaned, steps_csv_10_participants_cleaned)

# Concatenate the DataFrames on 'pid' and maintain the row count
try:
    # Concatenate the DataFrames along columns (axis=1), aligned on 'pid'
    concatenated_df = pd.concat([sleep_csv_10_participants_cleaned, steps_csv_10_participants_cleaned], axis=1)

    # Avoid duplicate 'pid' columns if any were added during concatenation
    concatenated_df = concatenated_df.loc[:, ~concatenated_df.columns.duplicated()]

    # Display the concatenated data
    print(concatenated_df)

    # Optionally, you can save the concatenated data to a CSV file
    concatenated_df.to_csv('concatenated_10_participants_steps_sleep.csv', index=False)
    print("Concatenated data for the first 10 participants has been saved as concatenated_10_participants_all.csv")
except MemoryError:
    print("Memory Error: Consider processing in smaller chunks or optimizing data types.")
```

```python
import pandas as pd
from sklearn.feature_selection import mutual_info_regression
import dcor
import pingouin as pg

# Step 1: Identify the columns containing 'avgefficiency'
avgefficiency_cols = [col for col in concatenated_df.columns if 'avgefficiency' in col.lower()]

print(f"Found 'avgefficiency' columns: {avgefficiency_cols}")

# Step 2: Drop rows with NaN values in the selected 'avgefficiency' columns
cleaned_df = concatenated_df.dropna(subset=avgefficiency_cols)

# Step 3: Select only numeric columns for correlation calculation
numeric_df = cleaned_df.select_dtypes(include=[float, int])

# Store results in a dictionary to save later
correlation_results = {}

# Spearman Correlation
spearman_corr = numeric_df.corr(method='spearman')
spearman_corr.to_csv('spearman_correlation.csv')
correlation_results['spearman'] = spearman_corr
print("Spearman correlation saved as 'spearman_correlation.csv'")

# Kendall Correlation
kendall_corr = numeric_df.corr(method='kendall')
kendall_corr.to_csv('kendall_correlation.csv')
correlation_results['kendall'] = kendall_corr
print("Kendall correlation saved as 'kendall_correlation.csv'")

# Mutual Information for 'avgefficiency' columns
mutual_info_scores = {}
for col in avgefficiency_cols:
    if col in numeric_df.columns:
        mi_scores = {}
        for predictor_col in numeric_df.columns:
            if predictor_col != col:
                mi_score = mutual_info_regression(numeric_df[[predictor_col]], numeric_df[col])
                mi_scores[predictor_col] = mi_score[0]
        mutual_info_scores[col] = mi_scores

# Convert mutual information scores to DataFrame and save
for col, scores in mutual_info_scores.items():
    mi_df = pd.DataFrame(list(scores.items()), columns=['Variable', 'MI_Score'])
    mi_df.to_csv(f'mutual_information_{col}.csv', index=False)
    print(f"Mutual information for {col} saved as 'mutual_information_{col}.csv'")

# Distance Correlation
distance_corr_results = {}
for col in avgefficiency_cols:
    if col in numeric_df.columns:
        dist_corr_scores = {}
        for predictor_col in numeric_df.columns:
            if predictor_col != col:
                dist_corr = dcor.distance_correlation(numeric_df[predictor_col], numeric_df[col])
                dist_corr_scores[predictor_col] = dist_corr
        distance_corr_results[col] = dist_corr_scores

# Convert distance correlation results to DataFrame and save
for col, scores in distance_corr_results.items():
    dist_corr_df = pd.DataFrame(list(scores.items()), columns=['Variable', 'Distance_Correlation'])
    dist_corr_df.to_csv(f'distance_correlation_{col}.csv', index=False)
    print(f"Distance correlation for {col} saved as 'distance_correlation_{col}.csv'")

# Hoeffding's D Test
hoeffding_results = {}
for col in avgefficiency_cols:
    if col in numeric_df.columns:
        hoeffding_scores = {}
        for predictor_col in numeric_df.columns:
            if predictor_col != col:
                hoeffding_d = pg.hoeffding(numeric_df[predictor_col], numeric_df[col])
                hoeffding_scores[predictor_col] = hoeffding_d['D'].values[0]
        hoeffding_results[col] = hoeffding_scores

# Convert Hoeffding's D results to DataFrame and save
for col, scores in hoeffding_results.items():
    hoeffding_df = pd.DataFrame(list(scores.items()), columns=['Variable', 'Hoeffding_D'])
    hoeffding_df.to_csv(f'hoeffding_{col}.csv', index=False)
    print(f"Hoeffding's D for {col} saved as 'hoeffding_{col}.csv'")
```
